###Grafico de densidad:

```python
sns.kdeplot(bd1['Year'],color="red")
```

Este grafico de densidad nos muestra algo muy parecido al gráfico de lineas, pero en este caso es la proporción de ventas históricas que han habido en cada año. Con esta gráfica podemos concluir que entre los años 2000 y 2010 han estados más del 50% de los videojuegos más vendidos.

###Joint plot:
Este gráfico nos es de vital ayuda ya que es una mezcla de los scatter plots o hexagonal plots con un histograma para cada eje. Estos gráficos nos muestran la misma historia de las gráficas anteriores, que las ventas de videojuegos en Europa y en Japón no están tan correlacionadas, por lo tanto son mercados con demandas distintas.

```python
sns.jointplot(x='JP_Sales', y='EU_Sales', data=bd1[bd1['Year'] > 2010].sample(40),color="green")
```

###Box plot:
Con esta gráfica podemos ver que los videojuegos más vendidos generalmente son valores atápicos, ya que más del 50% de los videojuegos (lo que vendría siendo la "caja" del boxplot) están en un rango de ventas diferentes para las plataformas de videojuegos.

```python
df = bd1[bd1.Platform.isin(bd1.Platform.value_counts().head(3).index)]
sns.boxplot(data=df,x='Platform',y='Global_Sales')
```

###Violin plot:
La gráfica de violín nos muestra información muy parecida al diagrama de caja, ya que lo que está graficando es la distribución de cada consola de videojuegos. Y en este caso podemos concluir lo mismo en cuanto a que los videojuegos que más se han vendido corresponden a valores atípicos.

```python
df1 = bd1[bd1.Platform.isin(bd1.Platform.value_counts().head(3).index)]
sns.violinplot(data=df1,x='Platform',y='Global_Sales')
```

###Pair plot:
Con el pair plot podemos ver las correlaciones entre las ventas de los videojuegos de Europa, Japon y Norte América. Y lo que podemos concluir es que las ventas en Estados Unidos y las ventas en Norte América están relacionadas, por lo que intuimos que si a un videojuego le va muy bien en ventas en Europa, también le iría muy bien en las ventas en Norte América.

```python
sns.pairplot(bd1[['NA_Sales', 'EU_Sales', 'JP_Sales']])
```

###Gráficas multi-variantes:

```python
sns.lmplot(x='NA_Sales', y='EU_Sales', hue='Genre', 
           data=bd1.loc[bd1['Genre'].isin(['Sports', 'Action'])].sample(40), 
           fit_reg=False)
```

Esta grafica nos permite ver que hay una relación entre las ventas de videojuegos de acción en Europa y en Norte America, por lo que suponemos que el genero de acción es muy popular en Norte America y en Europa. Mientras que los videojuegos de deportes estan más dispersos y cargados del lado de Europa, por lo que suponemos que es una preferencia solo de los clientes europeos.

```python
sport = plt.hist(bd1[bd1["Genre"] == "Adventure"].Global_Sales.head(30),bins=30,fc = (1,0,0,0.5),label = "Aventura")
act = plt.hist(bd1[bd1["Genre"] == "Shooter"].Global_Sales.head(30),bins=30,fc = (0,1,0,0.5),label = "Disparos")
plt.legend()
plt.xlabel("Ventas globales")
plt.ylabel("Frecuencia")
plt.title("Histograma")
plt.show()
```

Esta gráfica nos permite comparar entre las ventas que han tenido los videojuegos del género de aventuras y el género de disparos. Lo que nos arroja es que los videojuegos de disparos han tenido un mejor papel en el mercado, ya que tienen una frecuencia más alta de videojuegos con volúmenes de ventas altos. 

```python
mdata = pd.melt(bd1,id_vars = "Genre",value_vars = ['JP_Sales'])
plt.figure(figsize = (20,8))
sns.boxplot(x ="variable", y = "value", hue="Genre",data= mdata.head(20))
plt.show()
```

Esta gráfica es muy interesante pues nos arroja información acerca de como se comportan en ventas los distintos tipos de generos de videojuegos en Japón. Y podemos concluir que Japón tiene distintas preferencias de videojuegos con respecto al mercado occidental, ya que los generos que mejores ventas han tenido son los juegos de rol, de plataformas y los puzzles. Y estos no han tenido tan buenas ventas en los otros grandes mercados del mundo.

###Mapa de calor:

```python
sns.heatmap(bd1.corr(),annot= True,linewidths=0.5,fmt = ".1f")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Mapa de correlación')
plt.show()
```

Lo que nos parece mas importante a destacar de esta gráfica son 2 cosas, la primera es que ahora sabemos que las ventas en Europa están muy correlacionadas con las ventas en Norte América. Y lo segundo que quisieramos destacar es que las ventas en Norte América estan muy relacionadas con las ventas globales, por lo que suponemos que Norte América es el mercado de videojuegos más grande del mundo.

####A pesar de eso, la gráfica ideal para cada caso particular será seleccionada dependiendo de la información que se desea transmitir y los propósitos de la misma. Por ejemplo:
####**Diagrama de violín:**
*   Mostrar los intervalos de confianza
*   Enseñar la forma de la distribución

####**Diagrama de Boxplot**
*   Identificar valores atípicos
*   Conocer de una forma cómoda y rápida como el 50% de los valores centrales se distribuyen
