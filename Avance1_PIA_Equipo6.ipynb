>>- **text_reviews_count**: Los datos que describe esta columna son mediciones con el numero total de criticas escritas que obtuvo el libro.

>>- **publication_date**: Los datos que describe esta columna son tiempos con la fecha en la que se publico el libro.

>>- **publisher**: Los datos que describe esta columna son textos largos con la editora que publico el libro.

>**d)** Hacer el ejemplo de clasificación de datos para que elijan acomodar los datos para que se facilite esta tarea.

```python
print(bada.language_code.unique())  #Buscamos los valores que se encuentran en la columna 'language_code' para poder clasificar nuestros datos entre los
                                    #códigos de lenguaje de cada libro

#Se eliminaron las columnas 'isbn' e 'isbn13' ya que se consideran innecesarias puesto que ya contamos con el no. de ID de cada libro
del(bada['isbn'])   
del(bada['isbn13'])

#Acomodamos nuestra tabla de manera que podamos ver los libros y a un lado su código de lenguaje
bada=bada.reindex(columns=['bookID','title','authors','language_code','ratings_count','average_rating'])
bada
```

>**e)** Cambiar nombres de columnas.

>>[Liga para DUDAS](http://exponentis.es/como-eliminar-columnas-de-un-dataframe-de-python-pandas)

```python
bada.rename(columns = {'bookID':'ID libro','title':'Título','authors':'Autores','language_code':'Código Lenguaje','num_pages':'Páginas',
                     'average_rating':'Rating prom','ratings_count':'Cant. Ratings','text_reviews_count':'Opiniones','publication_date':'Fecha-publicacón',
                     'publisher':'Editorial'}).head(2)
#Esta funcion nos ayuda a cambiar el nombre de una columna con la siguiente forma: 'Columna':'Nuevo nombre'
#Aqui cambiamos los nombres de las columnas a su traducción al español.
```

```python
#AGRUPACIÓN 1

print("Valores de la columna 'Código-Lenguaje': \n",bada.language_code.unique())  #Para revisar los valores que se encuentran en la columna 'Código-Lenguaje'

agrup_1=bada[bada['language_code']=='en-US']  #Se agrupan los libros que su código de lenguaje sea inglés americano.
agrup_1.to_csv('Solo_EN-US.csv', index=False)

agrup_1
```

```python
#AGRUPACIÓN 2
#Se sigue la misma dinámica de la agrupación 1 pero para los libros cuyo código de lenguaje sea Inglés.

agrup_2 = bada[bada['language_code']=='eng'] 
agrup_2.to_csv('Solo_ENG.csv', index=False)

agrup_2
```
