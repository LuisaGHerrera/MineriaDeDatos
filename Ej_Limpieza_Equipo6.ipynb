> **d)**   Hacer el ejemplo de clasificación de datos para que elijan acomodar los datos para que se facilite esta tarea.

```python
print(bd.Type.unique())  #Buscamos los valores que se encuentran en la columna 'Type' para poder clasificar nuestros datos entre las Apps
                         #que son gratis y las que son de paga.

#Acomodamos nuestra tabla de manera que podamos ver las App y a un lado si son gratis o de paga (y el precio)
bd=bd.reindex(columns=['App','Type','Price','Genres','Rating','Size'])
bd
```

> **e)**   Cambiar nombres de columnas.

```python
bd.rename(columns = {'Category':'Categoría','Reviews':'Opiniones','Size':'Tamaño','Installs':'Descargas','Price':'Precio','Content Rating':'Clasificación',
          'Genres':'Género','Last Updated':'Última actualización','Current Ver':'Versión Actual','Android Ver':'Versión Android','Type':'Tipo'})
#Esta funcion nos ayuda a cambiar el nombre de una columna con la siguiente forma: 'Columna':'Nuevo nombre'
```

> **f)**   Guardar archivos en agrupaciones.

```python
#AGRUPACIÓN 1

print(bd.Type.unique())   #Para revisar los valores que se encuentran en la columna 'Type'

agrup_1=bd[bd['Type']=='Paid']  #Se aclara como se agruparán los datos, en este caso que cumpla que en la columna 'Type' diga 'Paid'. De paga.
agrup_1.to_csv('Solo_DE_PAGO.csv', index=False)

print(agrup_1)   #Esta impresión es auxiliar por si se quiere ver cuantas filas y columnas se agruparon.
```

```python
#AGRUPACIÓN 2
#Se sigue la misma dinámica de la agrupación 1 pero para las App Gratis

print(bd.Type.unique())

agrup_2 = bd[bd['Type']=='Free']
agrup_2.to_csv('Solo_GRATIS.csv', index=False)

print(agrup_2)
```
